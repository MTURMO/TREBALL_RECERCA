```python
def remove_punctuation ( text ):
    return re.sub('[%s]' % re.escape(string.punctuation), '', text)

def remove_punctuation2 (text):
    for char in "'-.,\n()":
        text=text.replace(char,' ')
    return text
```

```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
```
